```python
import pandas as pd
import numpy as np
from datetime import datetime
```

# Load Raw Data

```python
price20 = pd.read_csv('data/Price_be_utc2020.csv')
price21 = pd.read_csv('data/Price_be_utc2021.csv')
price22 = pd.read_csv('data/Price_be_utc2022.csv')
load20 = pd.read_csv('data/Load_be2020.csv')
load21 = pd.read_csv('data/Load_be2021.csv')
load22 = pd.read_csv('data/Load_be2022.csv')


price = pd.concat([price20,price21,price22])
load = pd.concat([load20,load21,load22])


```

```python
sum_forecast = 0
sum_actual = 0
actual = []
forecast = []

for index, row in load.iterrows(): 
    sum_forecast = sum_forecast + row[1]
    sum_actual = sum_actual + row[2]
    if (index+1) % 4 == 0:
        ave_for = sum_forecast / 4
        ave_act = sum_actual / 4
        actual.append(ave_act)
        forecast.append(ave_for)
        sum_actual = 0
        sum_forecast = 0
```

```python
date = pd.date_range(start=datetime.strptime('01/01/2020 00','%d/%m/%Y %H'),end=datetime.strptime('31/12/2022 23','%d/%m/%Y %H'),freq='H')
d = {'date':date,'price':price['Day-ahead Price [EUR/MWh]'], 'ActualLoad':actual,'ForecastLoad':forecast}
data = pd.DataFrame(d)
```

```python
weekday = []
month = []
hours = []
season = []
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
seasons = {
    1: 'winter',
    2: 'winter',
    3: 'spring',
    4: 'spring',
    5: 'spring',
    6: 'summer',
    7: 'summer',
    8: 'summer',
    9: 'fall',
    10: 'fall',
    11: 'fall',
    12: 'winter'
}

Months = {
    1:'Jan', 2: 'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'
}
timeslots= {
    0: "night", 1: "night", 2:"night", 3: "night", 4:"night",5:"night",
    6: "morning",7: "morning",8: "morning",9: "morning",10: "morning",11: "morning",
    12:"afternoon",13:"afternoon",14:"afternoon",15:"afternoon",16:"afternoon",17:"afternoon",
    18: "evening",19 : "evening",20 : "evening",21: "evening",22: "evening",23: "evening"
}
for index,row in data.iterrows():
    hours.append(timeslots[row['date'].hour])
    weekday.append(days[row['date'].weekday()])
    season.append(seasons[row['date'].month])
    month.append(Months[row['date'].month])


data['weekday'] = weekday
data['season'] = season
data['period'] = hours
#data['month'] = month

data.to_csv('data/data_be.csv',index=False)
```
